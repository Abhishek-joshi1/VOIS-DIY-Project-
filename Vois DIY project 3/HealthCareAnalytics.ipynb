```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'documents/1719219834-DoctorVisits - DA (1).csv'
doctor_visits_df = pd.read_csv(file_path)
```

```python
# Drop the 'Unnamed: 0' column as it is just an index
doctor_visits_df = doctor_visits_df.drop(columns=['Unnamed: 0'])
```

```python
# Convert categorical columns to appropriate data types
categorical_columns = ['gender', 'private', 'freepoor', 'freerepat', 'nchronic', 'lchronic']
for column in categorical_columns:
    doctor_visits_df[column] = doctor_visits_df[column].astype('category')
```

```python
missing_values = doctor_visits_df.isnull().sum()

print("Missing values in each column:\n", missing_values)
```

```python
# Descriptive statistics for numeric columns
descriptive_stats = doctor_visits_df.describe()
print("\nDescriptive statistics for numeric columns:\n", descriptive_stats)
```

```python
# Descriptive statistics for categorical columns
categorical_stats = doctor_visits_df[categorical_columns].describe()
print("\nDescriptive statistics for categorical columns:\n", categorical_stats)
```

```python
# Distribution of doctor visits
plt.figure(figsize=(10, 6))
sns.countplot(data=doctor_visits_df, x='visits', hue='visits', palette='viridis', legend=False)
plt.title('Distribution of Doctor Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of health status
plt.figure(figsize=(10, 6))
sns.countplot(data=doctor_visits_df, x='health', hue='health', palette='viridis', legend=False)
plt.title('Distribution of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of income by number of visits
plt.figure(figsize=(10, 6))
sns.boxplot(data=doctor_visits_df, x='visits', y='income',hue='visits', palette='viridis',legend=True)
plt.title('Income Distribution by Number of Doctor Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Income')
plt.show()
```

```python
# Boxplot of age by number of visits
plt.figure(figsize=(10, 6))
sns.boxplot(data=doctor_visits_df, x='visits', y='age', hue = "visits",palette='viridis')
plt.title('Age Distribution by Number of Doctor Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Age')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = doctor_visits_df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
